## TiC-Tac-Toe Game Using Python

```python

#-----------Variables----------

#Creating a Board for the game
board = ["-", "-", "-",
        "-", "-", "-",
        "-", "-", "-",]

#If game going
game_still_going = True

#Who Won? or Tie?
winner = None

#Whos Turn?
current_player = 'X'

#-----------Functions----------#

def display_board():
  print(board[0] +  " | " +board[1]+ " | " +board[2])
  print(board[3] +  " | " +board[4]+ " | " +board[5])
  print(board[6] +  " | " +board[7]+ " | " +board[8])


#Creating playing option & developing the game logic
def play_game():
  
  #display initial board
  display_board()

  #-----Loop containing game logic-----#

  #While the game is going.
  while game_still_going:

    #Turn of a player:
    handle_turn(current_player)

    #check if the game has ended
    check_if_game_over()
    
    #Flipping the player
    flip_player()

  #The game has ended
  if winner == 'X' or winner == 'O':
    print("Boom! " +  winner  + " Won. Congratulations!" )
  elif winner == None:
    print("Opps! It's a Tie.")
      

#Handle a single turn
def handle_turn(player):
  #Who's Turn?
  print( player + "'s Turn'")

  #Inputing Position
  position = input("Choose a Position from 1-9:")

  #for invalid inputs
  valid = False
  while not valid:
    
    #not using if loop cause it catches error just once
    while position not in ["1","2","3","4","5","6","7","8","9"]:
      position = input("Invalid Input.Choose a Position from 1-9:")
    
    #Making it Int, as input is Str type in 'board' variable
    #Changing the index value 1-9 for better understanding of the player.
    position = int(position)-1 

    #For Overlapping Issues
    if board[position] == '-':
      valid = True  #Exits the invalid checking loop 
    else:
      print("Position is used previously.Choose another place")

  #inputing value at the certain position
  board[position ]= player
  display_board() #displaying again


def check_if_game_over():
  check_for_winner()
  check_if_tie()

def check_for_winner():

  #calling up the variable outside the function
  global winner
  
  #check Rows
  row_winner = check_rows()
  #check Columns
  column_winner = check_colums()
  #check Diagonals
  diagonal_winner = check_diagonals()

  if row_winner:
    winner = row_winner
  elif column_winner:
    winner = column_winner
  elif diagonal_winner:
    winner = diagonal_winner
  else:
    winner = None
  return


def check_rows():
  
  #calling global variable to check
  global game_still_going

  #check if rows are same
  row_1 = board[0] == board[1] == board[2] != '-'
  row_2 = board[3] == board[4] == board[5] != '-'
  row_3 = board[6] == board[7] == board[8] != '-'

  #If any row is same then game stops
  if row_1 or row_2 or row_3:
    game_still_going = False

  #return the winner (X or O)
  if row_1:
    return board[0]
  elif row_2:
    return board[3]
  elif row_3:
    return board[6]

  return

  
def check_colums():

  #calling global variable to check
  global game_still_going

  #check if columns are same
  columns_1 = board[0] == board[3] == board[6] != '-'
  columns_2 = board[1] == board[4] == board[7] != '-'
  columns_3 = board[2] == board[5] == board[8] != '-'

  #If any column is same then game stops
  if columns_1 or columns_2 or columns_3:
    game_still_going = False

  #return the winner (X or O)
  if columns_1:
    return board[0]
  elif columns_2:
    return board[1]
  elif columns_3:
    return board[2]
    
  return


def check_diagonals():
  
  #calling global variable to check
  global game_still_going

  #check if columns are same
  diagonal_1 = board[0] == board[4] == board[8] != '-'
  diagonal_2 = board[2] == board[4] == board[6] != '-'

  #If any column is same then game stops
  if diagonal_1 or diagonal_2:
    game_still_going = False

  #return the winner (X or O)
  if diagonal_1:
    return board[0]
  elif diagonal_2:
    return board[2]
    
  return
  
def check_if_tie():
  global game_still_going
  if "-" not in board:
    game_still_going = False
  return


def flip_player():

  #Calling Global variable
  global current_player

  #If current player is X change it to O
  if current_player == 'X':
    current_player = 'O'
  #If current player is O change it to X
  elif current_player == 'O':
    current_player = 'X'
  return

play_game()
```
